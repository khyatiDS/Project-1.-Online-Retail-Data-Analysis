### Data Cleaning!

Checking for missing data

```python

df = pd.read_csv(r"C:\Users\laptop\OneDrive\Documents\Python\data\OnlineRetail.csv", encoding="ISO-8859-1")
df.isnull().sum()
```

Handling Missing Data

```python
df['Description'] = df['Description'].fillna("No Description")
df.dropna(subset=['CustomerID'])
```

Removing null values from data

```python
df.dropna()
```

Dropping duplicates

```python
df.drop_duplicates()
```

Data information

```python
df.info()
```

Checking the Data types

```python
print(df.dtypes)
```

Removing Invalid quantity and prices

```python
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

Creating total price column

```python
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

## Data Visualization

Convering InvoiceDate to Datetime 

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
```

Sales trend over 24 hours

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], errors='coerce')
df['Hour'] = df['InvoiceDate'].dt.hour

# Total revenue per hour
hourly_revenue = df.groupby('Hour')['TotalPrice'].sum()

# Visuzlisation
plt.figure(figsize=(12,6))
plt.plot(hourly_revenue.index, hourly_revenue.values, marker='o', color='orange', linewidth=2)

plt.title("Sales Trend Over 24 Hours", fontweight='bold', fontsize=16)
plt.xlabel("Hour of Day", fontweight='bold', fontsize=12)
plt.ylabel("Revenue", fontweight='bold', fontsize=12)

plt.xticks(range(0,24), fontweight='bold')
plt.yticks(fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Daily Revenue

```python
daily_revenue = df.groupby(df['InvoiceDate'].dt.date)['TotalPrice'].sum()
print(daily_revenue.head())
```

Visualization of Daily Revenue

```python
plt.figure(figsize=(10,5))
plt.plot(daily_revenue.index, daily_revenue.values)

plt.title("Daily Revenue Over Time", fontweight='bold', fontsize=16)
plt.xlabel("Date", fontweight='bold', fontsize=14)
plt.ylabel("Revenue", fontweight='bold', fontsize=14)

plt.xticks(rotation=45)
plt.show()
```

Monthly Revenue

```python
monthly_revenue = df.groupby(df['InvoiceDate'].dt.to_period('M'))['TotalPrice'].sum()
```

Visualisation of Monthly Revenue

```python
df['Month'] = df['InvoiceDate'].dt.month
monthly_revenue = df.groupby('Month')['TotalPrice'].sum()
monthly_revenue.plot(kind='bar', color='purple', figsize=(10,5))
plt.title("Monthly Revenue", fontweight='bold')
plt.xlabel("Month", fontweight='bold')
plt.ylabel("Revenue", fontweight='bold')
plt.show()
```

In the month of December the highest revenue is generated.

### Data Exploration!

#### Which products are selling the most?

```python
# Ensure TotalPrice exists
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Total quantity sold per product
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
print("Top 10 Products by Units Sold:")
print(top_products)


plt.figure(figsize=(12,6))
top_products.plot(kind='bar', color='red')
plt.title("Top 10 Products by Units Sold", fontweight='bold', fontsize=16)
plt.xlabel("Product", fontweight='bold', fontsize=12)
plt.ylabel("Units Sold", fontweight='bold', fontsize=12)
plt.xticks(rotation=45, fontweight='bold')
plt.show()
```

The top selling product with "PAPER CRAFT, LITTILE BIRDIE" and "MEDIUM CERAMIC TOP STORAGE JAR" leading the sales.

#### Who are the most valuable customers?

```python
# Total revenue per customer
top_customers = df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(10)
print("Top 10 Customers by Revenue:")
print(top_customers)

# Visualization
plt.figure(figsize=(12,6))
top_customers.plot(kind='bar', color='blue')
plt.title("Top 10 Customers by Revenue", fontweight='bold', fontsize=16)
plt.xlabel("Customer ID", fontweight='bold', fontsize=12)
plt.ylabel("Revenue", fontweight='bold', fontsize=12)
plt.xticks(rotation=45, fontweight='bold')
plt.show()
```

#### Which countries contribute the most revenue?

```python
# Revenue per country
country_revenue = df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False).head(10)
print("Top 10 Countries by Revenue:")
print(country_revenue)

# Visualization
plt.figure(figsize=(12,6))
country_revenue.plot(kind='bar', color='purple')
plt.title("Top 10 Countries by Revenue", fontweight='bold', fontsize=16)
plt.xlabel("Country", fontweight='bold', fontsize=12)
plt.ylabel("Revenue", fontweight='bold', fontsize=12)
plt.xticks(rotation=45, fontweight='bold')
plt.show()
```

```python
country_product = df.groupby(['Country','Description'])['Quantity'].sum().sort_values(ascending=False)
print(country_product.head(10))
```

#### Is high selling Product also generate high revenue?

```python
product_stats = df.groupby('Description').agg({'Quantity':'sum', 'TotalPrice':'sum'})
plt.figure(figsize=(10,6))
plt.scatter(product_stats['Quantity'], product_stats['TotalPrice'], alpha=0.5)
plt.xlabel("Total Quantity Sold", fontweight='bold')
plt.ylabel("Total Revenue", fontweight='bold')
plt.title("Revenue vs Quantity Sold per Product", fontweight='bold')
plt.show()
```

Most products generate low sales and low revwnue, while a few stand out as high revenue outliers - some from large volumes of low-priced itmes, others from fewer but high-priced premium products
